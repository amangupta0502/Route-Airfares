## Linear Regression Model with Lasso

```python
alphas = np.linspace(1, 0.001, 15)
alphas
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(allfeatures, target, test_size=0.25, random_state=42)
  
rmse_list = []
r2_list = []
model_list = []

for a in alphas:
    lasso = Lasso(alpha=a, max_iter=1000)
    lasso.fit(X_train, Y_train)
    pred = lasso.predict(X_test)
    
    r2_list.append(lasso.score(X_train,Y_train))
    rmse = mean_squared_error(Y_test, pred)**0.5;
    rmse_list.append(rmse)
    model_list.append(lasso)
    print("Alpha",a,"RMSE",rmse)
    

lasso_df = pd.DataFrame(zip(alphas, rmse_list, r2_list), columns=['Alpha', 'RMSE', "R2"])
lasso_df
```

```python
# Displaying the best model that has the lowest RMSE value
lasso_df.sort_values(by=['RMSE', 'R2'], ascending=[True, False]).head(1) 
```

## Using  variance threshold Feature Selection method

```python
target_feature = df.FARE
all_features = df.drop("FARE", axis=1)

vt = VarianceThreshold(threshold=0.15)
vt.fit(allfeatures)
pd.DataFrame({'variance': vt.variances_, 'select_feature': vt.get_support()}, index=allfeatures.columns)
df_vt = allfeatures.iloc[:,vt.get_support()]
df_vt

X_train, X_test, y_train, y_test = train_test_split(df_vt, target_feature,test_size=0.25, random_state=42)
```

```python
for model in models:
    pipeline = Pipeline([("transform", PolynomialFeatures(degree=2,include_bias=False,interaction_only=True)), ("scaler", RobustScaler()), ("regressor", model)])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    list_r2.append(pipeline.score(X_test, y_test))
    list_rmse.append(mean_squared_error(y_test, y_pred)**0.5)
```

```python
pipeline_result = pd.DataFrame(zip(name,list_r2,list_rmse), columns=['Regressor', 'R2', 'RMSE'])
```

```python
# Sorting according to RMSE values.
pipeline_result.sort_values(by="RMSE") 
```
